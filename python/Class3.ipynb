# Sequence sum
1/1! + 2/2! + 3/3! + ...

```python
n = int(input('enter n'))

result = 0
fact = 1

for i in range(1,n+1):
  fact = fact * i
  result = result + i/fact

print(result)
```

# Nested Loops

```python
for i in range(1,5):
  for j in range(1,5):
    print(i,j)
```

```python
# Pattern 1
# ***
# ****
# ***
```

```python
rows = int(input('enter number of rows'))

for i in range(1,rows+1):
  for j in range(1,i+1):
    print('*',end='')
  print()
```

```python
# Pattern 2
# 1
# 121
# 12321
# 1234321
```

```python
rows = int(input('enter number of rows'))

for i in range(1,rows+1):
  for j in range(1,i+1):
    print(j,end='')
  for k in range(i-1,0,-1):
    print(k,end='')

  print()
```

# Loop Control Statement

```python
for i in range(1,10):
  if i == 5:
    break
  print(i)
```
